## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
df = dataset.copy()
#Replace Null values with 'Not present'
df['name'].fillna('Not present',inplace = True)
df['host_name'].fillna('Not present',inplace = True)
df['last_review'].fillna('Not present',inplace = True)
df['reviews_per_month'].fillna('Not present',inplace = True)
df.head()
````

````python
#New column to store total minimum price
df["Minimum_stay_price"] = df['price'] * df['minimum_nights']
df["Minimum_stay_price"].head()
````

````python
ngroup_min_price = df.groupby('room_type')['Minimum_stay_price'].mean().reset_index().rename(columns={'Minimum_stay_price':'Min_price'}).sort_values('Min_price',ascending=False)
ngroup_min_price.head()
````

````python
#group_by roomt_type and count
room_type_and_count = df['room_type'].value_counts()
room_type_and_count.head()
````

````python
#Group by neighbourhood_group and average price
avg_price_by_ngroup = (
    df.groupby('neighbourhood_group')['Minimum_stay_price']
    .mean()
    .reset_index()
    .rename(columns={'Minimum_stay_price': 'avg_price'})
    .sort_values('avg_price')
)
avg_price_by_ngroup.head()
````

````python
#Group by neighbourhood and average price
avg_price_by_neighbourhood = df.groupby('neighbourhood')['price'].mean().reset_index().rename(columns={'price':'avg_price'}).sort_values('avg_price')
avg_price_by_neighbourhood.head()
````

````python
#Group by neighbourhood and roomtype
room_type_by_neighbourhood = df.groupby(['neighbourhood','room_type'])['price'].mean().reset_index().rename(columns={'price':'avg_price'})
room_type_by_neighbourhood.head()
````

````python
#room_count_by_neighbourhood.head()
room_count_by_neighbourhood = df.groupby(['neighbourhood', 'room_type']).size().reset_index(name='room_count')
room_count_by_neighbourhood.head()
````

````python
# Chart - 13 visualization code
# Top 10 neighborhoods by listing count
top_neigh = df['neighbourhood'].value_counts().nlargest(10)
total = top_neigh.sum()
percentages = (top_neigh / total) * 100

plt.figure(figsize=(10,6))
ax = sns.barplot(x=top_neigh.values, y=top_neigh.index, palette='mako')

# Add percentage text to bars
for i, (count, percent) in enumerate(zip(top_neigh.values, percentages)):
    ax.text(
        count + 5,                      # X position (slightly outside the bar)
        i,                              # Y position (bar index)
        f'{percent:.1f}%',              # Text
        va='center',                    # Vertical alignment
        fontsize=10
    )

plt.title('Top 10 Neighborhoods by Number of Listings')
plt.xlabel('Number of Listings')
plt.ylabel('Neighborhood')
plt.tight_layout()
plt.show()
````

##### 1. Why did you pick the specific chart?

The barplot lists the number of average listing acquired in each neighbourhood.

#### Chart - 15 - Pair Plot

````python
# Pair Plot visualization code


# Select a subset of numerical columns
pairplot_df = df[['price', 'minimum_nights', 'number_of_reviews', 'availability_365']].copy()

# Optional: cap extreme values to avoid distortion
pairplot_df = pairplot_df[(pairplot_df['price'] <= 500) & (pairplot_df['minimum_nights'] <= 30)]

# Create the pair plot
sns.pairplot(pairplot_df)
plt.suptitle('Pair Plot of Airbnb Listing Metrics', y=1.02)
plt.tight_layout()
plt.show()
````

##### 1. Why did you pick the specific chart?

A pair plot gives a comparison of various column of the dataset and hence helps analyze the trend through bars and plots.
